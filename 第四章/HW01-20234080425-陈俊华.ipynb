```python
import torch

# ---------------------- 1. 前向计算 ----------------------
def forward(x, w1, w2):
    a = x * w1
    b = a + w2
    L = b ** 2
    return a, b, L

# 给定输入
x = 2
w1_val = 1.5
w2_val = 0.5

a, b, L = forward(x, w1_val, w2_val)
print("前向计算结果：")
print(f"a = {a}, b = {b}, L = {L}")

# ---------------------- 2. 手动反向传播 ----------------------
# 计算梯度
dL_db = 2 * b
dL_dw2 = dL_db * 1
dL_da = dL_db * 1
dL_dw1 = dL_da * x

print("\n手动反向传播梯度：")
print(f"∂L/∂w1 = {dL_dw1}")
print(f"∂L/∂w2 = {dL_dw2}")

# ---------------------- 3. PyTorch 自动微分验证 ----------------------
# 定义可求导的张量
w1_torch = torch.tensor([w1_val], requires_grad=True)
w2_torch = torch.tensor([w2_val], requires_grad=True)
x_torch = torch.tensor([x], requires_grad=False)

# 前向计算
a_torch = x_torch * w1_torch
b_torch = a_torch + w2_torch
L_torch = b_torch ** 2

# 反向传播
L_torch.backward()

print("\nPyTorch 自动微分梯度：")
print(f"∂L/∂w1 = {w1_torch.grad.item()}")
print(f"∂L/∂w2 = {w2_torch.grad.item()}")
```
